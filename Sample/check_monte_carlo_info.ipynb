```python
import pylhe
import numpy as np

import ROOT
delphes_path = '/usr/local/Delphes-3.4.2'
ROOT.gROOT.ProcessLine(f'.include {delphes_path}/')
ROOT.gROOT.ProcessLine(f'.include {delphes_path}/external/')
ROOT.gInterpreter.Declare(f'#include "{delphes_path}/classes/DelphesClasses.h"')
ROOT.gSystem.Load(f'{delphes_path}/install/lib/libDelphes')
```

```python
lhe_path = './MG5/TRSM_420_280/Events/run_01_decayed_1/unweighted_events.lhe.gz'

lhe_file = pylhe.readLHE(lhe_path)

event_list = [event for event in lhe_file]
    
print("total events: {} ".format(len(event_list)))
```

```python
index = np.random.randint(low=1, high=len(event_list))

print("{:^12}{:^9}{:^12}{:^12}{:^12}{:^12}{:^12}{:^12}{:^12}".format("#","id","mother1","mother2","e","px","py","pz","status"))

for i, element in  enumerate(event_list[index].particles):
    print("{:^12}{:^9.0f}{:^12.0f}{:^12.0f}{:^12.3f}{:^12.3f}{:^12.3f}{:^12.3f}{:^12.0f}".format(i, element.id,element.mother1,element.mother2,element.e,element.px,element.py,element.pz,element.status))
```

```python
root_path = './MG5/TRSM_420_280/Events/run_01_decayed_1/tag_1_delphes_events.root'
root_file = ROOT.TFile(root_path)
tree = root_file.Get('Delphes')

nevent = tree.GetEntries()

tree.GetEntry(24)
print(f'{"Index":<5} {"PID":<5} {"M1":<5} {"M2":<5} {"D1":<5} {"D2":<5} {"PT":<10} {"Eta":<10} {"Phi":<10} {"Mass":<10}')
for particle_id, particle in enumerate(tree.Particle):
    print(f'{particle_id:<5} {particle.PID:<5} {particle.M1:<5} {particle.M2:<5} {particle.D1:<5} {particle.D2:<5} {particle.PT:<10.5f} {particle.Eta:<10.5f} {particle.Phi:<10.5f} {particle.Mass:<10.5f}')
```
