```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_auc(ax, m3_m2_list, file_pattern, label):
    AUCs = []
    for m3_m2 in m3_m2_list:
        m3_m2 = m3_m2.replace('/', '_')
        df = pd.read_csv(file_pattern.format(m3_m2))
        AUCs.append(df['AUC'])
    ax.plot(m3_m2_list, AUCs, marker='o', label=label)
```

```python
m3_m2_list = ['420/280', '500/275', '500/300', '520/325']

fig, ax = plt.subplots(figsize=(5, 4))

plot_auc(ax, m3_m2_list, './classification_performance/gghhh_bsm_{}/gghhh-4pT40_4b-test-DenseNN_abs_pairing.csv', 'Dense-NN w/ $|m - m_h|$')
plot_auc(ax, m3_m2_list, './classification_performance/gghhh_bsm_{}/gghhh-4pT40_4b-test-DenseNN_chi2_pairing.csv', 'Dense-NN w/ $(m - m_h)^2$')
plot_auc(ax, m3_m2_list, './classification_performance/gghhh_bsm_{}/gghhh-4pT40_4b-test-DenseNN_4b_SPANet_pairing.csv', 'Dense-NN w/ SPANet')
plot_auc(ax, m3_m2_list, './classification_performance/gghhh_bsm_{}/gghhh-4pT40_4b-test-SPANet_classifier.csv', 'SPANet classifier')

ax.set_title(f'$4b$ datasets')
ax.set_xlabel('$m_{h_3}/m_{h_2}$ [GeV]')
ax.set_ylabel('AUC')
ax.legend(frameon=False, loc='center right')

plt.savefig('./figures/classification_performance-4b_datasets.pdf', dpi=300, facecolor='White', bbox_inches='tight')
plt.show()
```
