```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_DNN_feature(file):
    results = np.load(file)

    results = {'dR1': results[:, 0],
            'dR2': results[:, 1],
            'dR3': results[:, 2],
            'rms_dR': results[:, 3],
            'dA_skew': results[:, 4],
            'HT': results[:, 5],
            'mhCostheta': results[:, 6],
            'eta_mhhh_fraction': results[:, 7],
            'sphericity': results[:, 8],
            'aplanarity': results[:, 9]}

    return results
```

```python
results_res = load_DNN_feature('./DNN_features/resonant-abs_pairing.npy')
```

```python
results_6b = load_DNN_feature('DNN_features/pp6b-abs_pairing.npy')
```

```python
# plot dR distribution
fig, ax = plt.subplots(1, 3, figsize=(15, 4))

for i in range(3):
    # normalize the histogram

    ax[i].hist(results_6b[f'dR{i+1}'], bins=20, range=(0, 4.5), histtype='step', density=True, label='Background')
    ax[i].hist(results_res[f'dR{i+1}'], bins=20, range=(0, 4.5), histtype='step', density=True, label=f'Signal')
    ax[i].legend(frameon=False)
    ax[i].set_xlabel('$\Delta R$')
    ax[i].set_ylabel('Normalized counts')
    ax[i].set_title(f'$\Delta R_{i+1}$ distribution')

plt.savefig('../figures/dR_distribution-abs_pairing.pdf', dpi=300, facecolor='White', bbox_inches='tight')
plt.show()
```

```python
# plot RMS of dR distribution
fig, ax = plt.subplots(1, 1, figsize=(5, 4))

ax.hist(results_6b['rms_dR'], bins=20, range=(0, 5), histtype='step', density=True, label='Background')
ax.hist(results_res['rms_dR'], bins=20, range=(0, 5), histtype='step', density=True, label='Signal')
ax.legend(frameon=False)
ax.set_xlabel('RMS of $\Delta R$')
ax.set_ylabel('Normalized counts')
ax.set_title('RMS of $\Delta R$ distribution')

plt.savefig('../figures/rms_dR_distribution-abs_pairing.pdf', dpi=300, facecolor='White', bbox_inches='tight')
plt.show()
```

```python
# plot skewness of dA distribution
fig, ax = plt.subplots(1, 1, figsize=(5, 4))

ax.hist(results_6b['dA_skew'], bins=20, range=(-1, 3.5), histtype='step', density=True, label='Background')
ax.hist(results_res['dA_skew'], bins=20, range=(-1, 3.5), histtype='step', density=True, label='Signal')
ax.legend(frameon=False)
ax.set_xlabel('Skewness of $\Delta A$')
ax.set_ylabel('Normalized counts')
ax.set_title('Skewness of $\Delta A$ distribution')

plt.savefig('../figures/dA_skew_distribution-abs_pairing.pdf', dpi=300, facecolor='White', bbox_inches='tight')
plt.show()
```

```python
# plot HT distribution
fig, ax = plt.subplots(1, 1, figsize=(5, 4))

ax.hist(results_6b['HT'], bins=20, range=(200, 1600), histtype='step', density=True, label='Background')
ax.hist(results_res['HT'], bins=20, range=(200, 1600), histtype='step', density=True, label='Signal')
ax.legend(frameon=False)
ax.set_xlabel('$H_{T, \mathrm{6jets}}$ (GeV)')
ax.set_ylabel('Normalized counts')
ax.set_title('$H_T$ distribution')

plt.savefig('../figures/HT_distribution-abs_pairing.pdf', dpi=300, facecolor='White', bbox_inches='tight')
plt.show()
```
